```python
# Shuffle original array vs permutation

from numpy import random

arr = [1, 2, 3, 4, 5]

print(arr)

random.shuffle(arr)
arr_permutated = random.permutation(arr)

print(f"After shuffling: {arr}")
print(f"After permutation: {arr_permutated}")
```

```python
# Random distribution with probability density function

from numpy import random

arr = random.choice(
    [2, 3, 3.5, 4, 4.5, 5],
    p = (0.1, 0.25, 0.25, 0.15, 0.15, 0.1),
    size = 5,
)
print(arr)
```

```python
# Normal/Gaussian distribution and its curve bell

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

arr = random.normal(
    loc = 5,
    scale = 20,
    size = 100,
)

sns.distplot(arr)
plt.show()
```

```python
# Binomal distribution for binary results (e.g. coin toss)

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

arr = random.binomial(
    n = 10,
    p = 0.5,
    size = 100,
)

sns.distplot(arr)
plt.show()
```

```python
# Multinomial distribution for multi-nomial outcomes (e.g. dice roll)

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

arr = random.multinomial(
    n = 6,
    pvals = [0.1, 0.25, 0.25, 0.15, 0.15, 0.1],
    size = 100,
)

sns.distplot(arr)
plt.show()
```

```python
# Uniform distribution where every event has equal chance of occuring

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

arr = random.uniform(
    low = 2,
    high = 5,
    size = 100,
)

sns.distplot(arr)
plt.show()
```

```python
# Poisson distribution for estimating how many times an event can happend

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

arr = random.poisson(
    lam = 2,
    size = 1000,
)

sns.distplot(arr)
plt.show()
```
