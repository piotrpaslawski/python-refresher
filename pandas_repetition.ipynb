```python
# Displaying specific column or row

import pandas as pd

data = {
    "Country": ["Poland", "Germany", "Russia", "Italy", "French"],
    "Capital City": ["Warsaw", "Berlin", "Moscow", "Rome", "Paris"],
}

df = pd.DataFrame(data)

print(df)

print(f"One column:\n{df['Country']}")
print(f"Two columns:\n{df[['Country', 'Capital City']]}")
print(f"Third row:\n{df.iloc[2]}")
```

```python
# Filtering, sorting and grouping elements

import pandas as pd

data = {
    "Name": ["John", "Emily", "Michael", "Jessica", "William"],
    "Age": [28, 24, 22, 30, 27],
    "City": ["New York", "Los Angeles", "New York", "Houston", "Miami"],
}

df = pd.DataFrame(data)

filtered = df[df["Age"] < 25]
sorted = df.sort_values(by="Age")

grouped = df.groupby("City")
average_age = grouped["Age"].mean()

print(f"Filtered:\n{filtered}")
print(f"Sorted:\n{sorted}")
print(f"Grouped with average age:\n{average_age}")
```

```python
# Deleting specific column or row

import pandas as pd

df = pd.DataFrame()

df["Letter"] = [chr(i) for i in range(65, 92)]
df["Code"] = [i for i in range(65, 92)]
df["Mistake"] = 1

df.drop(df.index[-1], inplace=True)
df.drop("Mistake", axis=1, inplace=True)

print(df)
```

```python
# Series, NaN elements and CSV/JSON files

import pandas as pd

pd.DataFrame(
    {
        "colA": pd.Series([1, 2, 3]),
        "colB": pd.Series([4, 5, 6, 7]),
    }
)

filename = "example.csv"
df.to_csv(filename, index=False)

new_df = pd.read_csv(filename)
print(new_df)
print(new_df.info())

filename = "example.json"
df.to_json(filename, orient="records")

new_df = pd.read_json(filename)
print(new_df)
print(new_df.info())
```

```python
# Example statistical analysis

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"

df = pd.read_csv(url)

print(f"Some first rows:\n{df.head()}")
print(f"Example row:\n{df.sample()}")
print(f"Basic statistic:\n{df.describe()}")
print(f"Averages in species:\n{df.groupby('species').mean()}")

sns.histplot(
    data=df,
    x="petal_length",
    bins=20,
)
plt.show()

sns.kdeplot(
    data=df,
    x="petal_length",
    hue="species",
    fill=True,
)
plt.show()
```
